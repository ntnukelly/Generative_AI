# KL Divergence是什麼?

KL Divergence是用來衡量「兩個機率分布有多不一樣」的一種方式


KL Divergence公式:

螢幕擷取畫面 2025-03-08 021835.png

如果P和Q完全一樣，KL散度就是0，表示模型預測得很準確

如果Q偏離P很多，KL散度就會變大，KL值越大代表模型的預測越不準確

所以訓練機器學習模型時，希望Q（模型預測）盡量接近P（真實數據分布）

# Cross Entropy是什麼？

交叉熵是用來衡量「真實分布P和預測分布Q之間的差異」的一種方法

如果模型預測得很準確（Q和P接近），交叉熵會比較小

但如果模型預測錯得離譜（Q和P差很多），交叉熵就會變大

# KL Divergence和Cross Entropy的適用情形：

當真實情況不是[1, 0, 0]、[0, 1, 0] 、[0, 0, 1]的時候適用KL Divergence

ex:給電腦一張貓咪圖，我想讓他辨識是貓還是狗照片，我會設定真實標籤P為貓（1），狗（0），這樣就會使用Cross Entropy

但如果像是做化學實驗，實驗結果不一定會是固定結果，在真實情況下出現A情況可能是70%，出現B情況是20%，出現C情況是10%，這樣就會用KL Divergence
，反之則用Cross Entropy

小結:

如果你是要「讓模型學到一個機率分布」 → 用KL Divergence

如果你是要「做分類」（找出最可能的類別） → 用Cross Entropy
