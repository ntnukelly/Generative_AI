# 畫函數圖形
畫f(x) = sin(πt/2)

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#在0到6之間產生1000個等間距的數值
t = np.linspace(0, 6, 1000)
x = np.sin(np.pi * t / 2)
y = t

#設定圖表大小、背景顏色、曲線顏色、曲線粗細
plt.figure(figsize=(8, 8), facecolor='white')
plt.plot(x, y, 'm', linewidth=3, label="Parametric Curve")

#圖上顯示公式(文字位置、字體大小、顏色、粗細)
plt.text(0, 5, "Curve: x = sin(πt/2), y = t", fontsize=14, ha='center', color='black', fontweight='bold')
```
